```python
import pandas as pd
import yfinance as yf
import numpy as np

ticker_symbol="RELIANCE.NS"

ticker=yf.Ticker(ticker_symbol)
data=ticker.history(period="2y")

n=int(input())
lower12=n-11
lower26=n-25
lower_Sig=n-8


def SMA(n): #SMA 20
  if n<19:
    sum=0
    for i in range(n+1):
      sum=sum+data["Close"].iloc[i]
    SMA_result=sum/n+1
  else:
    sum=0
    for i in range(n-19, n+1):
      sum=sum+data["Close"].iloc[i]
    SMA_result=sum/20
  return SMA_result

def EMA12(n):
  if lower12<0:
    EMA_result=0
  else:
    if n==0:
      EMA_result=SMA(0)
    else:
      if n==lower12:
        EMA_result=SMA(lower12) #Before func call: def lower12=n-11.
      else:
        Prev=EMA12(n-1)
        EMA_result=(2/13)*(data["Close"].iloc[n]-Prev)+Prev

  return EMA_result

def EMA26(n):
  if lower26<0:
    EMA_result=0
  else:
    if n==0:
      EMA_result=SMA(0)
    else:
      if n==lower26:
        EMA_result=SMA(lower26) #Bef func call: def lower26=n-25
      else:
        Prev=EMA26(n-1)
        EMA_result=(2/27)*(data["Close"].iloc[n]-Prev)+Prev

  return EMA_result

def MACD(n):
  if lower26<0:
    MACD_result=0
  else:
    EMA12_result=EMA12(n)
    EMA26_result=EMA26(n)
    MACD_result=EMA12_result-EMA26_result
  return MACD_result

def Signal(n):
  if n<34:
    Signal_result=0
  else:
    Prev=MACD(n-1)
    if n==lower_Sig:
      Signal_result=MACD(lower_Sig)
    else:
      Signal_result=(2/10)*(MACD(n)-Prev)+Prev
  return Signal_result

def RSI(n): #RSI 14
  if n<13:
    RSI_result=0
  else:
    Tot_Gain=0
    Tot_Loss=0
    Diff=0
    for i in range(n-13,n):
      Diff=data["Close"].iloc[i+1]-data["Close"].iloc[i]
      if Diff>0:
        Tot_Gain=Tot_Gain+Diff
      else:
        Tot_Loss=Tot_Loss-Diff
      Av_Gain=Tot_Gain/14
      Av_Loss=Tot_Loss/14
      if Av_Loss==0:
        RSI_result=100
      else:
        RSI_result=100-(100/(1+Av_Gain/Av_Loss))
  return RSI_result

def Stoch_perK(n):
  if n<13:
    percentK=0
  else:
    s=np.zeros(14)
    for i in range(n-13,n+1):
      s[i-n+13]=data["Close"].iloc[i]
    Min=np.min(s)
    Max=np.max(s)
    percentK=(data["Close"].iloc[n]-Min)/(Max-Min)*100
  return percentK

def Stoch_perD(n):
  if n<16:
    percentD=0
  else:
    sum=0
    for i in range(n-2,n+1):
      sum=sum+Stoch_perK(i)
    percentD=sum/3
  return percentD

def BB_MidBand(n):
  if n<19:
    Mid_Band=0
  else:
    Mid_Band=SMA(n)
  return Mid_Band

def BB_UpperBand(n):
  if n<19:
    Upper_Band=0
  else:
    s=np.zeros(20)
    for i in range(n-19,n+1):
      s[i-n+19]=data["Close"].iloc[i]
    Std=np.std(s)
    Upper_Band=BB_MidBand(n)+2*Std
  return Upper_Band

def BB_LowerBand(n):
  if n<19:
    Lower_Band=0
  else:
    s=np.zeros(20)
    for i in range(n-19,n+1):
      s[i-n+19]=data["Close"].iloc[i]
    Std=np.std(s)
    Lower_Band=BB_MidBand(n)-2*Std
  return Lower_Band


print(SMA(n), EMA12(n), EMA26(n), MACD(n), Signal(n), RSI(n), Stoch_perK(n), Stoch_perD(n), BB_MidBand(n), BB_UpperBand(n), BB_LowerBand(n))


```
